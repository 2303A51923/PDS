```python
import pandas as pd

df = pd.read_csv("healthcare_dataset.csv")
print("Dataset before cleaning:")
print(df.head(10))
invalid_ages = df[(df['Age'] < 0) | (df['Age'] > 120)]
print("\nInvalid Ages Found:")
print(invalid_ages)
df_cleaned = df[(df['Age'] >= 0) & (df['Age'] <= 120)]

print("\nDataset after cleaning:")
print(df_cleaned.head(10))
df_cleaned.to_csv("healthcare_dataset_cleaned.csv", index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
df = pd.read_csv("bank.csv")
print("Original Dataset (first 5 rows):")
display(df.head())

print("\nAvailable Columns:", df.columns)
income_column = "balance"
df['Income_Normalized'] = MinMaxScaler().fit_transform(df[[income_column]])
df['Income_Standardized'] = StandardScaler().fit_transform(df[[income_column]])

print("\nDataset After Scaling (first 5 rows):")
display(df[[income_column, "Income_Normalized", "Income_Standardized"]].head())
df.to_csv("bank_scaled.csv", index=False)
print("\n✅ Scaled dataset saved as 'bank_scaled.csv'")
```
